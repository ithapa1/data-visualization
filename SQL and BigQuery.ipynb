Now, that I've been verified, I can begin tinkering with the dataset. I'm going to look at the 'austin bikeshare.bikeshare trips' dataset and see how many entries there are, as well as the longest journey from 'durration mintues' and the average time for the trip.

```python
%%bigquery --project data-visualization-338803
SELECT 
  COUNT(*) as total_rows
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
```

```python
%%bigquery --project data-visualization-338803
SELECT 
  MAX(duration_minutes) 
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
```

```python
%%bigquery --project data-visualization-338803
SELECT 
  AVG(duration_minutes) 
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
```

```python
%%bigquery --project data-visualization-338803
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
LIMIT 5
```
