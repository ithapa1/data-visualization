Here, I am going to import the data from the london fire brigade.

Now, lets count the number of late night false alarms in between 10pm to 5am of the next day

```python
%%bigquery --project data-visualization-338803
SELECT  COUNT(*) as number_of_late_night_false_alarms
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
WHERE incident_group = "False Alarm" AND (hour_of_call >= 22 OR hour_of_call <= 4)
```

Here, I am going to find out the average number of pumps attending in the london fire brigade.

```python
%%bigquery --project data-visualization-338803
SELECT  incident_group, AVG(num_pumps_attending) as avg_pumps_attending
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
GROUP BY incident_group
```

Now, I am going to figure out the maximum that is largest number of pumps of the london fire brigade service calls.

```python
%%bigquery --project data-visualization-338803
SELECT  incident_group, MAX(num_pumps_attending) as Max_pumps_attending
FROM `bigquery-public-data.london_fire_brigade.fire_brigade_service_calls`
GROUP BY incident_group
```
