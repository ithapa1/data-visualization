```python
import pandas as pa
import seaborn  as sns
df = pa.read_csv('https://raw.githubusercontent.com/nurfnick/Data_Viz/main/Activity_Dataset_V1.csv')
df.head()
```

```python
df.calories.plot.box()
```

```python
df.boxplot(column = 'calories', by = 'workout_type',rot = 60, layout= None)
```

```python
df.loc[df["workout_type"]== "Trail Run"]
```

```python
by_Range = df.groupby("workout_type")
by_Range.get_group("Trail Run")
```

```python
by_Range.boxplot(subplots='Trail Run', column='calories', fontsize=15, rot=45, grid=True, ax=None, figsize= '15,15', layout=None, sharex=False, sharey=True, backend=None)
```

```python
df.boxplot(column = 'calories', by = 'workout_type',vert = False)
```

```python
by_Range.boxplot(column = 'calories',by = 'workout_type',vert = True, rot=None, figsize=(15,15), sharex=True, sharey=True)
```

```python
ax = sns.boxplot(data = df, x = 'workout_type', y = 'calories')
ax.set_xticklabels(ax.get_xticklabels(),rotation = 60)
```

```python
sns.boxplot( y = df.workout_type, x = df.calories)
```

```python
ax = sns.boxplot(data = df, y = 'workout_type', x = 'calories')
```
