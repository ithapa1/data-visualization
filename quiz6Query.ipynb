Here, I am going to upload all the data related to the parkland as questioned ask to answer.

```python
%%bigquery --project data-visualization-338803
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations`
WHERE property_type = 'parkland'
LIMIT 10
```

Now, I am going to pull all the information related to the sidewalk.  

```python
%%bigquery --project data-visualization-338803
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations`
WHERE property_type = 'sidewalk'
LIMIT 10
```

Here, I am going to join the two data from different stations and its trips.

```python
%%bigquery --project data-visualization-338803
WITH t as(
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_stations`
WHERE property_type = 'parkland'
)
SELECT *
FROM t as table1 JOIN `bigquery-public-data.austin_bikeshare.bikeshare_trips` as table2 ON table1.name = table2.start_station_name
limit 10
```

```python
%%bigquery --project data-visualization-338803
SELECT type.station_id,type.name, type.property_type,trips.start_station_name,trips.end_station_name
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips` as trips LEFT OUTER JOIN `bigquery-public-data.austin_bikeshare.bikeshare_stations` as type
          on trips.start_station_name = type.name
LIMIT 40
```

I calculated and pull out all the possible station id, property type, name of the station, start station name and the end station name.
