```python
sns.relplot(data=df, x = 'Goals', y= 'Assists' )
plt.title('A scatterplot of goals and assists')
```

Here, you can see the graph of scatter plots of Barcelona's goals and assists is shown above.

```python
def BarcelonaGoals(row):
  return row.Goals
def BarcelonaAssists(row):
  return row.Assists
```

We utilize a function in the preceding code that takes a row of data and returns the sum of goals and assists. To compute the scatter of goals and assists, we use the str.contains function and lambda

```python
df2 = df[df['club'].str.contains('Barcelona')]

df['FCB_goals'] = df2.apply(lambda row: BarcelonaGoals(row), axis=1)

df['FCB_assists'] = df2.apply(lambda row: BarcelonaAssists(row), axis=1)
```

```python
df['FCB_goals'] = df['FCB_goals'].fillna(0).astype(int)
df['FCB_goals'] = df['FCB_goals'].astype('int')
df['FCB_assists'] = df['FCB_assists'].fillna(0).astype(int)
df['FCB_assists'] = df['FCB_assists'].astype('int')
```

```python
df[df['club'].str.contains('Barcelona')]
```

With str.contains Barcelona loaded, the aforementioned dataset is ready to plot the scatter plot graph. So, I am going to show you the scatter plot of the goals in the following graph which are as follows:

```python
sns.relplot(data = df, x = 'Goals', y = 'Assists', color = 'green', style = 'FCB_goals', size = 'FCB_assists')

plt.title('A scatterplot of goals vs assist')
```

Now, you can see the dataset of the goals and assists of the Football Club Barcelona in the scatter plot.
