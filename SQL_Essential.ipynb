Now that Authentication is done, I will be looking at the data and play around. I am going to plot all the data related to the austin bikeshare trips.

```python
%%bigquery --project data-visualization-338803
SELECT *
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
LIMIT 5
```

now, i am going to plot the station name and average time.

```python
%%bigquery --project data-visualization-338803
SELECT AVG(duration_minutes) as avg_time, start_station_name
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
Order by avg_time
```

here, lets put the data related to no_trips and start station name.

```python
%%bigquery --project data-visualization-338803
SELECT COUNT(start_station_name) as No_Trip, start_station_name
FROM `bigquery-public-data.austin_bikeshare.bikeshare_trips`
GROUP BY start_station_name
Order by No_Trip
```
