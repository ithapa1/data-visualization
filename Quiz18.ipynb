```python
df = pa.read_csv("https://raw.githubusercontent.com/nurfnick/Data_Viz/main/AB_NYC_2019.csv")
df.head()
```

```python
edlevels = sorted(df.neighbourhood_group.unique())

edlevels
```

```python
eddict = {}

for i in range(5):
  eddict[edlevels[i]] = i

eddict
```

```python
df.groupby('neighbourhood_group').host_id.agg('count')
```

```python
df.neighbourhood_group.map(eddict)
```

```python
df.neighbourhood_group = df.neighbourhood_group.map(eddict)

df.head()
```

At last, I transformed the numerical data separately because I didn't know how to convert it in the same table and it was difficult. I used for-loop to assign a numerical value to each neighborhood group, then used the groupby function to group and to convert them.
