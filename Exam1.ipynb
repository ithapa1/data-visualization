the following dataset has 21938 rows in total.

```python
%%bigquery --project data-visualization-338803
SELECT COUNT(unique_key) as unique_key, fare
FROM `bigquery-public-data.chicago_taxi_trips.taxi_trips`
GROUP BY fare
Order by unique_key
#total rows
```

here, the following are the  average trip total based on payment type.

```python
%%bigquery --project data-visualization-338803
SELECT AVG(trip_total) as trip_total, payment_type
FROM `bigquery-public-data.chicago_taxi_trips.taxi_trips`
GROUP BY payment_type
Order by trip_total
```

Now, I am testing the non-zero recorded time that is above the average time.

```python
%%bigquery --project data-visualization-338803
SELECT COUNT (*) as Avg_NonZero_Trip
FROM `bigquery-public-data.chicago_taxi_trips.taxi_trips`
where trip_seconds > 0 And trip_seconds > (select avg(trip_seconds) as average_trip FROM `bigquery-public-data.chicago_taxi_trips.taxi_trips`)
```
